```python
import matplotlib.pyplot as plt
from mtcnn.mtcnn import MTCNN
import cv2
```

```python
img=cv2.imread("image.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_new=img.copy()
plt.imshow(img_new)
plt.axis("off")
plt.show()
```

```python
detector=MTCNN()
faces=detector.detect_faces(img_new)
for face in faces:            ## 6 faces in pic 
    print(face)
```

```python
for face in faces:
    x,y,w,h = face['box']
    # create the shape
    green=(0,255,0)
    red=(255,0,0)
    cv2.rectangle(img_new,(x, y),(x+w,y+h),green,5)
    for key, value in face['keypoints'].items():
        # create and draw dot
        cv2.circle(img_new,value,2,red,5)
plt.imshow(img_new)
plt.show()
```

```python
plt.figure(figsize=(20,20))
for idx,face in enumerate(faces):
    # get coordinates
    x,y,w,h = face['box']
    plt.subplot(1,len(faces),idx+1)
    plt.axis('off')
    plt.imshow(img[y:y+h, x:x+w])
plt.show()
```
